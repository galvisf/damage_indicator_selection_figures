#### Inputs for moving bin method

#### function for the moving bin, by number of instances in each bin

```python
def fragility_by_moving_bin_by_number(results_filename, edp_results, edp_values, bin_size, im_type, p_values):
    #
    # INPUTS
    #     results_filename =  
    #     edp_results      = 
    #     edp_values       = 
    #     bin_size         =
    #     im_type          = 
    #     p_values         = 
    #
    #
    
    gm_ids = edp_results.index
    n_gms = len(gm_ids)
    scales = edp_results.columns
    df = pd.DataFrame(index=gm_ids, columns=scales)

    edp_results = edp_results.to_numpy()
    edp_results_list = np.sort(np.ravel(edp_results))
    n_results = len(edp_results_list)
    
    key = 'intact_results/ida/collapse_fragilities'
    intact_median = pd.read_hdf(results_filename, key).loc[im_type, 'Median']
    intact_beta = pd.read_hdf(results_filename, key).loc[im_type, 'Beta']

    gm_scale_group = 'mainshock_damage_results/' + gm_ids[0] + '/' + str(scales[0]) + 'Col'
    key = gm_scale_group + '/ida/collapse_intensities'
    columns = pd.read_hdf(results_filename, key=key).columns

    cols = ['Median', 'Beta', 'Min EDP', 'Max EDP', 'N Damaged Instances', 'N <=DI']
    p_cols = list()
    for p in p_values:
        p_cols.append('P(Col|IM_P(Col,Intact)='+'{0:.0f}'.format(100*p)+'%)')
        
    fragilities = pd.DataFrame(columns=cols+p_cols, dtype='float64') # create DataFrame object
    for i_edp, i in zip(edp_values, range(len(edp_values))):
#         edp_min = max(i_edp - edp_buffer, 0)
        edp_min = i_edp
    
        idx = np.searchsorted(edp_results_list, i_edp)
        
        idx_max = idx + bin_size - 1
        if idx_max >= n_results:
            idx
            edp_max = edp_results_list[-1]
        else:
            edp_max = edp_results_list[idx_max]
            
        idx_min = idx - bin_size
        if idx_min <= 0:
            idx_min = 0
            edp_min = 0
        else:
            edp_min = edp_results_list[idx_min]

        [gm_idx, scale_idx] = np.where((edp_results >= edp_min) & (edp_results <= edp_max))
        df.iloc[gm_idx, scale_idx] = i_edp
        n_instances = len(gm_idx)

        total_collapse_intensities = pd.DataFrame(columns=columns, dtype='float64')
        for i_gm, i_scale in zip(gm_idx, scale_idx):
            gm_id = gm_ids[i_gm]
            scale_name = str(scales[i_scale]) + 'Col'

            gm_scale_group = 'mainshock_damage_results/' + gm_id + '/' + scale_name
            key = gm_scale_group + '/ida/collapse_intensities'
            collapse_intensities = pd.read_hdf(results_filename, key=key)
            total_collapse_intensities = pd.concat([total_collapse_intensities, collapse_intensities], axis=0)

        if n_instances > 0:
            fragilities.loc[i_edp, ['Median', 'Beta']] = compute_ida_fragility(
                total_collapse_intensities[im_type], plot=False)
            for p, p_name in zip(p_values, p_cols):
                im = stats.lognorm(intact_beta, scale=intact_median).ppf(p)
                fragilities.loc[i_edp, p_name] = np.sum(total_collapse_intensities[im_type] <= im) / (n_gms * n_instances)
        
        fragilities.loc[i_edp, 'Min EDP'] = edp_min
        fragilities.loc[i_edp, 'Max EDP'] = edp_max
        fragilities.loc[i_edp, 'N Damaged Instances'] = n_instances
        fragilities.loc[i_edp, 'N <=DI'] = idx

    return fragilities
```

#### read damage indicators and organize as a matrix of mainshock gm vs scale

```python
def read_damage_indicator(damage_indicator_matrix, edp_type, n_gms, scales):
    
    n_scales = len(scales)
    
    # Create ground motion list
    gm_list = []
    for gm_i in range(n_gms):
        gm_list.append( 'GM' + str(gm_i + 1))
 
    # Reshape edp_results
    edp_results_array = damage_indicator_matrix[edp_type].to_numpy()
    edp_results = np.zeros([n_gms, n_scales])
    
    i = 0
    for gm_i in range(n_gms):
        for scale_i in range(n_scales):
            edp_results[gm_i, scale_i] = edp_results_array[i]
            i += 1
    
    edp_results = pd.DataFrame(data=edp_results, index=gm_list, columns=scales)  
    
    return edp_results
```

#### collecting the collapse intensity points for the scatter plot

```python
def aftershock_collapse_intensities_by_edp(results_filename, edp_results, im_type):
    
    gm_ids = edp_results.index
    scales = edp_results.columns

    n_gms = len(gm_ids)
    n_scales = len(scales)

    edp = np.zeros((n_gms * n_scales * n_gms))
    aftershock_col_intensity = np.zeros_like(edp)
    
    i = 0
    for gm_id in gm_ids:
        for scale in scales:
            start_i = i * n_gms
            end_i = start_i + n_gms
            edp[start_i:end_i] = edp_results.loc[gm_id, scale]

            key = 'mainshock_damage_results/' + gm_id + '/' + str(scale) + 'Col' + '/ida/collapse_intensities'
            collapse_intensities = pd.read_hdf(results_filename, key)[im_type]
            aftershock_col_intensity[start_i:end_i] = collapse_intensities
            
            i = i + 1

    edp_col_intensity = pd.DataFrame({'EDP': edp, 'IM': aftershock_col_intensity})
    
    return edp_col_intensity
```

#### wrapper for the functions

```python
def plot_and_save_DI_values(results_filename, fragility_filename, edp_results, edp_value_type, n_steps, bin_type, \
                            bin_sizes, bin_overlaps, im_type, p_values):
    edp_min = max(np.min(edp_results.to_numpy()), 1e-12)
    edp_max = np.max(edp_results.to_numpy())
    
    if edp_value_type == 'log':
        edp_values = np.exp(np.linspace(np.log(edp_min), np.log(edp_max), n_steps))
    elif edp_value_type == 'linear':
        edp_values = np.linspace(edp_min, edp_max, n_steps)
        
    # for bin wdith    
    edp_buffers = np.array([0.025, 0.05, 0.1, 0.15])
    edp_buffers = edp_buffers
    edp_buffers = edp_buffers * edp_max
    
#     bin_sizes = bin_sizes[3:]
    
#     for edp_buffer in edp_buffers:
    for j, bin_size in enumerate(bin_sizes):
        
        if bin_type == 'by_number':
            if edp_value_type == 'dynamic':
                
                edp_results_list = np.sort(np.ravel(edp_results.to_numpy()))                
                
                n_results = len(edp_results_list)
                
#                 # Idx to move the bin every bin_size damage instances
#                 idx = np.arange(bin_size, n_results, bin_size).astype('int')
                
                # Idx to move the bin every damage instance
                idx = np.arange(bin_size, n_results, bin_overlaps[j]).astype('int')
                
                if idx[-1] > n_results - bin_size:
                    idx[-1] = n_results - bin_size
                edp_values = edp_results_list[idx]
            fragilities = fragility_by_moving_bin_by_number(results_filename, edp_results, edp_values, bin_size, im_type, p_values)
        elif bin_type == 'by_width':
            fragilities = fragility_by_moving_bin_by_width(results_filename, edp_results, edp_values, edp_buffer, im_type, p_values)
        
        # Assemble and save dataframe with fragilities from moving bin method
        intact_row = pd.DataFrame(index=[0], columns=fragilities.columns)
        intact_row['Median'] = intact_median
        intact_row['Beta'] = intact_beta
        fragilities_complete = pd.concat([intact_row, fragilities])
        fragilities_complete.to_hdf(fragility_filename, key=edp_type)
        
        # Data for plots
        edp_col = aftershock_collapse_intensities_by_edp(results_filename, edp_results, im_type)
        edp_col['kappa'] = edp_col['IM'] / intact_median
        
        if plot_burton:
            if im_type == 'Sa(T1)':
                fragility_type = 'ida_sa_t1'
            else:
                fragility_type = 'ida_sa_avg'
            edp_median = k_by_damage_instance(results_filename, edp_results, fragility_type, None)
            edp_median['IM'] = edp_median['kappa'] * intact_median

        
#         fig = plt.figure(figsize=(15, 10))
#         current_ax = plt.subplots(1,1)
        fig,current_ax = plt.subplots(1,1,figsize=(9.5, 7))

        _ = current_ax.scatter(edp_col['EDP'], edp_col[y_value], color='lightgray', s=1, zorder=-5)
        if plot_burton:
            _ = current_ax.scatter(edp_median['EDP'], edp_median[y_value], color='dimgray', s=3, zorder=-3)
        
        if bin_type == 'by_width':
            _ = current_ax.errorbar(edp_max/2, max(edp_col[y_value]), xerr=edp_buffer, xlolims=True, capsize=3, capthick=1, marker='o', color='k')
    
        p_value_list = [0.8,0.5,0.2]
        p_im = (stats.lognorm(intact_beta, scale=intact_median).ppf([p_value_list[0], p_value_list[-1]]))
        p_im = [[p_im[0] - intact_median], [intact_median - p_im[1]]]
        median = intact_median
        if y_value == 'kappa':
            median = median / intact_median
            p_im = p_im / intact_median
        _ = current_ax.errorbar(edp_values[0], median, yerr=p_im, capsize=3, capthick=1, marker='o', color='k', zorder=5)


        color = 'tab:blue'
        marker = [None, 'o', None]
        
        for p, i in zip(p_value_list, range(len(p_value_list))):
            p_im = (stats.lognorm(fragilities['Beta'], scale=fragilities['Median']).ppf(p))
            if y_value == 'kappa':
                p_im = p_im / intact_median
            _ = current_ax.plot(fragilities.index, p_im, color=color, marker=marker[i], label='P(Collapse|DI) = '+'{0:.0f}'.format(100*p)+'%')

        xlabel_prefix = 'Damage Indicator:\n'
        if edp_type == '$SDR_{peak}$':
            xlabel = xlabel_prefix + 'Peak Story Drift Ratio'
        elif edp_type == '$RSDR_{peak}$':
            xlabel = xlabel_prefix + 'Residual Story Drift Ratio'
        elif edp_type == 'mainshock_sa_avg':
            xlabel = xlabel_prefix + 'Mainshock Sa$_{avg}$'
        elif edp_type == 'mainshock_sa_ratio':
            xlabel = xlabel_prefix + 'Mainshock Sa$_{ratio}$'
        elif edp_type == 'DI hinges':
            xlabel = xlabel_prefix + 'DI hinges'
        elif edp_type[0:3] == 'dsr':   
            xlabel = xlabel_prefix + edp_type
        else:
            if residual_type == 'avg':
                xlabel = xlabel_prefix + 'Average residual\n' + edp_type
            elif residual_type == 'area_under':
                xlabel = xlabel_prefix + 'Area under residual\n' + edp_type
            elif residual_type == 'roof':
                xlabel = xlabel_prefix + 'Residual at Roof\n' + edp_type
            
        _ = current_ax.set_xlabel(xlabel)
        if not xscale == 'linear':
            _ = current_ax.set_xscale(xscale)
            _ = current_ax.xaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda y, _: '{:.2g}'.format(y)))
        else:
            _ = current_ax.set_xlim(left = edp_min)                        
                                                
        if y_value == 'IM':
            intensity_ylabel = 'Collapse Intensity\n' + 'Sa$_{avg}$(T$_1$), [g]'
        elif y_value == 'kappa':
            intensity_ylabel = 'Normalized\nCollapse Intensity\n' + '$Sa_{col} / \widehat{Sa}_{col, Intact}$'
        _ = current_ax.set_ylabel(intensity_ylabel)
        _ = current_ax.set_ylim(bottom=0)
        _ = current_ax.legend(bbox_to_anchor=(1, 0.5), loc='center left')
        if bin_type == 'by_width':
            title = 'Bin Limits: + ' + '{0:.1g}'.format(edp_buffer)
        elif bin_type == 'by_number':
            title = 'Bin Size: +/-' + '{0:.0f}'.format(bin_size) + ' Damage Instances & Overlap of ' + str(bin_overlaps[j]) + ' damage instances'
        _ = current_ax.grid('on', color='dimgray', linestyle='--')
        _ = plt.title(title)
        _ = plt.show()
        
        
        fig, ax = plt.subplots(1,2, figsize=(12,5))

        current_ax = ax[0]
        _ = current_ax.scatter(edp_col['EDP'], edp_col[y_value], color='lightgray', s=1, zorder=-5)
        xlim = np.array([max(min(edp_col['EDP']), edp_min), max(edp_col['EDP'])])
        
        if plot_burton:
            _ = current_ax.scatter(edp_median['EDP'], edp_median[y_value], color='dimgray', s=3, zorder=-3)

        
        if bin_type == 'by_width':
            _ = current_ax.errorbar(edp_max/2, max(edp_col[y_value]), xerr=edp_buffer, xlolims=True, capsize=3, capthick=1, marker='o', color='k')


        for p, i in zip(p_values, range(len(p_values))):

            im = stats.lognorm(intact_beta, scale=intact_median).ppf(p)
            p_im = (stats.lognorm(fragilities['Beta'], scale=fragilities['Median']).cdf(im))
            if y_value == 'kappa':
                im = im / intact_median

            current_ax = ax[1]
            label = r'IM$_{P_{col,Intact}=' + '{0:.0f}'.format(100*p) + '\%}$'
            _ = current_ax.plot(fragilities.index, 100 * p_im, color='C'+str(i), label=label)
            _ = current_ax.plot(xlim, [100*p]*2, color='C'+str(i), linestyle='--')

            current_ax = ax[0]
            label = r'IM$_{P_{col,Intact}=' + '{0:.0f}'.format(100*p) + '\%}$'
            _ = current_ax.plot(xlim, [im]*2, color='C'+str(i), label=label)
            
        current_ax = ax[1]
        _ = current_ax.plot(fragilities.index, fragilities['N <=DI']/(44*4/100), color='darkgray', zorder=-1, label='CDF of Damage')
            
        _ = current_ax.set_xlabel(xlabel)

        current_ax = ax[0]
        _ = current_ax.set_xlabel(xlabel)
        if not xscale == 'linear':
            _ = current_ax.set_xscale(xscale)
            _ = current_ax.xaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda y, _: '{:.2g}'.format(y)))
        _ = current_ax.set_xlim(xlim)
        
        _ = current_ax.set_ylabel(intensity_ylabel)
        _ = current_ax.set_ylim(bottom=0)
        if bin_type == 'by_width':
            title = 'DI Bin Limits:\n+ ' + '{0:.1g}'.format(edp_buffer)
        elif bin_type == 'by_number':
            title = 'DI Bin Size:\n+/-' + '{0:.0f}'.format(bin_size) + ' Damage Instances'
        _ = current_ax.set_title(title)

        current_ax = ax[1]
        _ = current_ax.set_xlabel(xlabel)
        if not xscale == 'linear':
            _ = current_ax.set_xscale(xscale)
            _ = current_ax.xaxis.set_major_formatter(mpl.ticker.FuncFormatter(lambda y, _: '{:.2g}'.format(y)))
        _ = current_ax.set_xlim(xlim)

        _ = current_ax.set_ylabel('P(Collapse | IM)')
        _ = current_ax.set_ylim(0,100)
        _ = current_ax.legend(bbox_to_anchor=(1,0.5), loc='center left')
        title = 'LogNormal Fit Probabilities'
        _ = current_ax.set_title(title)
        _ = current_ax.grid('on', color='lightgray')


        plt.tight_layout(rect=[0,0,1,0.93])
        _ = plt.show()
```
