#### INPUTS for figure

```python
fragilty_func_filename = '../1_Data_postprocessing/kappa_function_parameters.h5'
damage_instance_filename = '../1_Data_postprocessing/Damage_Instances_per_building.h5'

# building ids for fragilities
bldg_ids = ['ID1008', 'ID1012', 'ID1014', 'ID1021'] 
# idx for the default building
bldg_idx = 3
# fundamental period of the buildings
bldg_periods = [0.95, 1.8, 2.14, 2.3] 
```

#### Colors per building

```python
color_palettes = ['Oranges', 'Greens', 'Blues', 'Greys']
```

```python
# Conditioning Damage indicator
edp_type = '$SDR_{peak}$'
x_limits = [0.001, 0.1]
edp_list = [0.01, 0.02, 0.03]
# edp_type = '$DI_{sdr}$'
# x_limits = [0.001, 1.0]
# edp_list = [0.01, 0.10, 0.15]

# Intact fragility
median = 0.41
beta = 0.21


fig, ax_list = plt.subplots(1,2, figsize=(12,4))

i = 0
ax = ax_list[i]
parameters = pd.read_hdf(fragilty_func_filename, key=bldg_ids[bldg_idx])
damage_instances = pd.read_hdf(damage_instance_filename, key=bldg_ids[bldg_idx])

x = damage_instances[edp_type]
y = damage_instances['$\kappa$']
y_label = 'Fraction of $\widehat{Sa}_{collapse}^{Intact}, \kappa$'
x_data_space = 'log'
color_specs = 'dimgray'
color_line = 'dimgray'
building_title = ''

plotDIvsk3Lin_scatter(edp_type, x, y, x_limits, parameters, y_label, x_data_space, ax, color_specs, color_line, building_title)

kappa_list = kappa_func(edp_type, edp_list, parameters)

n_colors = len(edp_list)+1
min_intensity = 0.35
max_intensity = 0.99
color_values = np.flip(np.linspace(min_intensity, max_intensity, n_colors))
color_palettes = [color_palettes[bldg_idx]] + [color_palettes[bldg_idx]] * (n_colors-1)
fragility_colors = [mpl.cm.get_cmap(color_palettes[i])(color_values[i]) for i in range(len(color_values))] 

_ = ax.scatter(edp_list, kappa_list, s=50, color=fragility_colors[1:], zorder=10, edgecolor='k')

i = 1
ax = ax_list[i]
linewidth = 2.5


y = np.linspace(0.001, 1, num=100)

label = 'Intact Building, \n$\kappa$: {0:.1f}'.format(1) + ', $\widehat{Sa}_{col}=$' + '{0:.2f}'.format(median)
x = stats.lognorm(beta, scale=median).ppf(y)
color = fragility_colors[0]
_ = ax.plot(x, 100 * y, label=label, linewidth=linewidth, color=color)
_ = ax.scatter(median, 50, s=50, color=color, zorder=10, edgecolor='k')

for i_edp, edp in enumerate(edp_list):
    
    kappa = kappa_list[i_edp]
    damaged_median = kappa * median
    
    if edp_type == '$SDR_{peak}$' or edp_type == '$RSDR_{peak}$':
        label = edp_type + '= ' + '{0:.2f}'.format(100*edp) + '%\n$\kappa$: {0:.2f}'.format(kappa)+', $\widehat{Sa}_{col}=$' + '{0:.2f}'.format(damaged_median)    
    else:
        label = edp_type + '= ' + '{0:.2f}'.format(edp) + '\n$\kappa$: {0:.2f}'.format(kappa)+', $\widehat{Sa}_{col}=$' + '{0:.2f}'.format(damaged_median)    
    
    x = stats.lognorm(beta, scale=damaged_median).ppf(y)
    
    color = fragility_colors[i_edp+1]
    _ = ax.plot(x, 100 * y, label=label, linewidth=linewidth, color=color)
    _ = ax.scatter(damaged_median, 50, s=50, color=color, zorder=10, edgecolor='k')
    
_ = ax.legend(bbox_to_anchor=(1.05, 0.5), loc='center left')
_ = ax.set_ylabel('P(Collapse|Sa$_{avg}$(T)) [%]')
_ = ax.set_ylim(0, 100)
_ = ax.grid(color='lightgray')
_ = ax.set_xlim(0.2, 0.5)

label_tag = 'Sa$_{avg}$(T='
_ = ax.set_xlabel(label_tag + '{0:.1f}'.format(bldg_periods[bldg_idx])+ ') [g]')
```

```python
# Conditioning Damage indicator
edp_type = '$SDR_{peak}$'
x_limits = [0.001, 0.1]
edp_list = [0.01, 0.02, 0.03]
# edp_type = '$DI_{sdr}$'
# x_limits = [0.001, 1.0]
# edp_list = [0.01, 0.10, 0.15]

# Intact fragility
median = 0.41
beta = 0.21


fig, ax_list = plt.subplots(1,1, figsize=(5,4))

ax = ax_list
linewidth = 2.5

y = np.linspace(0.001, 1, num=100)

label = 'Intact Building, \n$\kappa$: {0:.1f}'.format(1) + ', $\widehat{Sa}_{col}=$' + '{0:.2f}'.format(median)
x = stats.lognorm(beta, scale=median).ppf(y)
color = fragility_colors[0]
_ = ax.plot(x, 100 * y, label=label, linewidth=linewidth, color=color)
_ = ax.scatter(median, 50, s=50, color=color, zorder=10, edgecolor='k')
_ = ax.plot([median, median], [0, 50], color='gray', linewidth=linewidth, linestyle='dotted')

for i_edp, edp in enumerate(edp_list):
    
    kappa = kappa_list[i_edp]
    damaged_median = kappa * median
    
    if edp_type == '$SDR_{peak}$' or edp_type == '$RSDR_{peak}$':
        label = edp_type + '= ' + '{0:.2f}'.format(100*edp) + '%\n$\kappa$: {0:.2f}'.format(kappa)+', $\widehat{Sa}_{col}=$' + '{0:.2f}'.format(damaged_median)    
    else:
        label = edp_type + '= ' + '{0:.2f}'.format(edp) + '\n$\kappa$: {0:.2f}'.format(kappa)+', $\widehat{Sa}_{col}=$' + '{0:.2f}'.format(damaged_median)    
    
    x = stats.lognorm(beta, scale=damaged_median).ppf(y)
    
    color = fragility_colors[i_edp+1]
    _ = ax.plot(x, 100 * y, label=label, linewidth=linewidth, color=color)
    _ = ax.scatter(damaged_median, 50, s=50, color=color, zorder=10, edgecolor='k')
    
_ = ax.set_ylabel('Probablity of collapse [%]')
_ = ax.plot([0, 0.5], [50, 50], color='gray', linewidth=linewidth, linestyle='dashed')
_ = ax.set_ylim(0, 100)
_ = ax.set_xlim(0.2, 0.5)

_ = ax.set_xlabel('Shaking Intensity')


_ = plt.tight_layout()
filename = 'damaged_fragilities.svg'
_ = plt.savefig(filename, bbox_inches='tight')
```
